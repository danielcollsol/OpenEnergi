# A. Data Extraction & Wrangling

### Import Libraries

```python
import numpy as np
import pandas as pd
from scipy.interpolate import lagrange 
import matplotlib.pyplot as plt
import traces
from datetime import timedelta, datetime
import requests
```

## 1. Power Measurements Extraction - CSV Source

```python
filepath_power_measurements = 'src/power_measurements_2019-11-25.csv'
```

```python
df_pm = pd.read_csv(filepath_power_measurements)
df_pm
```

```python
# Convert Time column in datetime datatype instead of string

def string_to_datetime(array):
    datetime_list = []
    for element in array:
        datetime_list.append(datetime.strptime(element, '%Y-%m-%d %H:%M:%S.%f0'))
    return datetime_list

time_col_list = df_pm['Time'].tolist()
df_pm['Time'] = string_to_datetime(time_col_list)
       
df_pm
```

```python
# Plot unevendly-spaced time series data

x = df_pm['Time']
y = df_pm['Value']

plt.plot(x,y,'x', mew=2, label='Samples')
plt.xlabel('Time')
plt.ylabel('Power Measurements [kW]')
plt.title("Power Mesurements of the asset abc123xyz, 25/11/2019")
plt.legend()
```

## 2. Regularize irregular time series with linear interpolation

```python
# https://stackoverflow.com/questions/25234941/python-regularise-irregular-time-series-with-linear-interpolation
# regularize data using linear interpolation

pair_time_value_list = list(zip(df_pm['Time'],df_pm['Value'])) # Prepare the traces input format

ts = traces.TimeSeries(pair_time_value_list)
reg = ts.sample(
    sampling_period=timedelta(minutes=30),
    start=datetime(2019, 11, 25, 0, 15),
    end=datetime(2019, 11, 26, 0),
    interpolate='linear',
)
```

```python
# retrieve regularized power values
reg_values = []
reg_datetime = []
reg_datetime_complete = []
for pairs in reg:
    reg_values.append(pairs[1])
    reg_datetime.append(pairs[0].strftime('%H:%M'))
    reg_datetime_complete.append(pairs[0])
```

```python
# Plot regularized power values
x = reg_datetime
y = reg_values

fig, ax = plt.subplots()
xticks = [1,9,17,25,33,41,47]
ax.set_xticks(xticks)

plt.plot(x,y,'x', mew=2, label='Data Points')
plt.xlabel('Time')
plt.ylabel('Reg. Power Measurements [kW]')
plt.title("Regularized Power Mesurements of the asset abc123xyz, 25/11/2019")
plt.legend()
```

```python
# Save values in dataframe
HalfHourId_list = list(range(1,49))
df_reg_power_measures = pd.DataFrame(list(zip(reg_datetime_complete,reg_values,HalfHourId_list)), columns =[ 'Datetime','Avg. Elec. Power [kW]', 'HalfHourId'])
df_reg_power_measures.head()
```

## 3. Carbon Intensity

### API Source

```python
# Request data through the API
headers = {
  'Accept': 'application/json'
}
 
r = requests.get('https://api.carbonintensity.org.uk/intensity/2019-11-25T00:00Z/2019-11-26T00:00Z', params={}, headers = headers) 

# Extract and reorganize data of interest (date, time and intensity) in a new dataframe

column_from = []
column_to = []
column_actual_intensity = []
for row in r.json()['data']:
    column_from.append(row['from'])
    column_to.append(row['to'])
    column_actual_intensity.append(row['intensity']['actual'])

dic = {}
dic['HalfHourId']= list(range(0,49))
dic['From [UTC (+00:00)]'] = column_from
dic['To [UTC (+00:00)]'] = column_to
dic['Intensity [gCO2/kWh]'] = column_actual_intensity


df_ci = pd.DataFrame(dic)

#delete those rows out of the date 2019/11/25
df_ci = df_ci[1:len(df_ci)]
print(df_ci.head())
```

### JSON Source

```python
filepath_carbon_intensity = 'src/carbon_intensity_2019-11-25.json'
```

```python
df_ci_raw = pd.read_json(filepath_carbon_intensity)
```

```python
# Inspect read data

print( '\n' + 'DataFrame retrieved:' + '\n')
print(df_ci_raw.head())
print( '\n' + 'Data dictionary sample:' + '\n')
print(df_ci_raw['data'][1])
```

```python
# Extract and reorganize data of interest (date, time and intensity) in a new dataframe

column_from = []
column_to = []
column_actual_intensity = []
for row in df_ci_raw.iterrows():
    column_from.append(row[1].get('data').get('from'))
    column_to.append(row[1].get('data').get('to'))
    column_actual_intensity.append(row[1].get('data').get('intensity').get('actual'))

dic = {}
dic['HalfHourId']= list(range(0,49))
dic['From [UTC (+00:00)]'] = column_from
dic['To [UTC (+00:00)]'] = column_to
dic['Intensity [gCO2/kWh]'] = column_actual_intensity


df_ci = pd.DataFrame(dic)

#delete those rows out of the date 2019/11/25
df_ci = df_ci[1:len(df_ci)]
print(df_ci.head())

```

# B. Calculations

```python
# Merge both dataframe in a new one for making the calculations
df_cf = df_ci.merge(df_reg_power_measures, how='inner')
df_cf.head()
```

```python
# Get average electrical consumption % carbon footprint
df_cf['Avg. Elec. Consumption [kWh]'] = 0.5 * df_cf['Avg. Elec. Power [kW]']
df_cf['Carbon Footprint [gCO2eq]'] = df_cf['Avg. Elec. Consumption [kWh]'] * df_cf['Intensity [gCO2/kWh]']

df_cf
```

```python
df_cf['Carbon Footprint [gCO2eq]'].sum()
```

# Bis - Minimum Threshold of Power Measurement COV Rule

```python
df_threshold = df_pm
print(df_threshold)
```

### Getting the minimal Threshold of the COV rule

```python
# Get the power measurements
pm_list = []
for row in df_threshold.iterrows():
    pm_list.append(row[1].get('Value'))
    
# Calculate the increment of the power measurements
new_list = []
i = 1
while i < len(pm_list):
    increment = abs(pm_list[i]-pm_list[i-1])
    new_list.append(increment)
    i= i +1
    
# Sort and inspect values
new_list.sort()  
print('The sorted list obtained is:' + '\n')
print(new_list[0:15], '\n')
print('The minimum threshold is:', round(new_list[15], 3), 'kW')
```

#### However, the graph of power measurements over time shows that COV rule threshold is not always the same ( difference of 750 kW between one point to another )
